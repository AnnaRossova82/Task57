```python
#Just for try - procees with color improve of another image
img = cv2.imread('sant2.JPG')
```

```python
b, g, r = cv2.split(img)
```

```python
b_corrected = cv2.equalizeHist(b)
g_corrected = cv2.equalizeHist(g)
r_corrected = cv2.equalizeHist(r)
```

```python
img_corrected = cv2.merge((b_corrected, g_corrected, r_corrected ))
```

```python
cv2.imshow('original', img)
cv2.imshow('corrected_colors', img_corrected)
cv2.waitKey(0)
#Get bright saturated new picture as a result
```

```python
plt.axis('off')
plt.imshow(img_corrected)
plt.show
```

```python
gamma = 0.5
```

```python
inv_gamma = 1.0 / gamma
table = np.array([((i/255.0)**inv_gamma)* 255 for i in np.arange(0, 256)]).astype('uint8')
```

```python
img_corrected2 = cv2.LUT(img, table)
```

```python
cv2.imshow('original', img)
cv2.imshow('corrected2_colors', img_corrected2)
cv2.waitKey(0)
#See more dark, clear details and bright saturated view 
```

```python
plt.axis('off')
plt.imshow(img_corrected2)
plt.show
```

```python
img3 = cv2.imread('birds.JPG')
plt.axis('off')
plt.imshow(img3)
plt.show
```

```python
# CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=4., tileGridSize=(8,8))
```

```python
lab = cv2.cvtColor(img3, cv2.COLOR_BGR2LAB)  # convert from BGR to LAB color space
l, a, b = cv2.split(lab)  # split on 3 different channels
```

```python
l2 = clahe.apply(l)  # apply CLAHE to the L-channel
```

```python
lab = cv2.merge((l2,a,b))  # merge channels
img4 = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)  # convert from LAB to BGR
#cv2.imshow('Increased contrast', img4)
plt.axis('off')
plt.imshow(img4)
plt.show
```

```python
from PIL import Image, ImageEnhance 
im = Image.open('birds.JPG') 
enhancer = ImageEnhance.Sharpness(im) 
factor = 1 
im_s_1 = enhancer.enhance(factor) 
im_s_1.save('birds.JPG')
factor2 = 0.05 
im_s_2 = enhancer.enhance(factor2) 
im_s_2.save('blurred_birgs.JPG')
factor3 = 2 
im_s_3 = enhancer.enhance(factor3) 
im_s_3.save('sharpened_bigds.JPG')
```

```python
plt.axis('off')
plt.imshow(im_s_1)
plt.show
```

```python
#image brightness enhancer 
enhancer = ImageEnhance.Brightness(im) 
factor = 1 #gives original image 
im_output = enhancer.enhance(factor) 
im_output.save('original-image.JPG') 
factor2 = 0.5 #darkens the image 
im_output2 = enhancer.enhance(factor2) 
im_output2.save('darkened-image.JPG') 
factor3 = 1.5 #brightens the image 
im_output3 = enhancer.enhance(factor3) 
im_output.save('brightened-image.JPG')
```
